```python
# range() IS DIFFERENCE BETWEEN MINIMUM VALUE AND MAXIMUM VALUE IN A SET OF NUMBERS .

# CALCULATE MEAN VALUE USING PYTHON .

a1 = [80,72,92,64,84,68,76,60,96,88]
a2 = max(a1) - min(a1)
print(a2)
```

```python
# CALCULATE MEAN VALUE USING NUMPY .

import numpy as np
b1 = np.array([80,72,92,64,84,68,76,60,96,88])
b2 = np.ptp(b1)
print(b2)
```

```python
# CALCULATE MEAN VALUE USING MATPLOTLIB .

import matplotlib.pyplot as plt
c = [13,16,11,15,14,12,13,15,11,14,12,12,14,11,13,12,11,13,11,12,11]
plt.hist ( c , bins=5 , edgecolor="black" , alpha=0.92 )

plt.axvline ( x=min(c) , color="orange" , linestyle="dashed" , linewidth=1.2 , label="MIN" )
plt.axvline ( x=max(c) , color="green" , linestyle="dashed" , linewidth=1.2 , label="MAX" )

plt.xlabel("VALUE")
plt.ylabel("FREQUENCY")
plt.title("HISTOGRAM PLOT")

plt.legend()
plt.show()
```
