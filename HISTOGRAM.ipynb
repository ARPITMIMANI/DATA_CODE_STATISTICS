```python
# CREATE A HISTOGRAM USING MATPLOTLIB .

import matplotlib.pyplot as plt
import random

a = [ random.randint(1,6) for _ in range(100) ]
plt.hist ( a , bins=range(1,8) , edgecolor="black" )

plt.title("HISTOGRAM")
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

c = [ random.randint(1,10) for _ in range(100) ]
plt.hist ( c , bins=range(1,12) , edgecolor="black" , width=0.8 )

plt.title("HISTOGRAM")
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

d = np.random.normal ( 100 , 10 , 200 )
plt.hist ( d , bins=20 , edgecolor="black" )

plt.title("HISTOGRAM")
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

d = np.random.normal ( loc=100 , scale=10 , size=200 )
plt.hist ( d , bins=20 , edgecolor="black" )

plt.title("HISTOGRAM")
plt.xlabel("X-AXIS")
plt.ylabel("Y-AXIS")
plt.show()
```
